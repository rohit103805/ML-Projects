# Smile Detection using OpenCV

## 1. Importing the Libraries

```python
import cv2
```

## 2. Importing the Cascades

```python
Eye_Cascade=cv2.CascadeClassifier('D:/ML_Projects/Smile_Detector_OpenCV/haarcascade_eye.xml')
Face_Cascade=cv2.CascadeClassifier('D:/ML_Projects/Smile_Detector_OpenCV/haarcascade_frontalface_default.xml')
Smile_Cascade=cv2.CascadeClassifier('D:/ML_Projects/Smile_Detector_OpenCV/haarcascade_smile.xml')
```

## 3. Drawing the Rectangles

```python
def Smile_detect(frame, gray):
    faces=Face_Cascade.detectMultiScale(gray, 1.5, 5)
    print(len(faces)) # No of Faces Detected .
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 0, 0), 3)
        sub_grey=grey[y:y+h, x:x+w]
        sub_frame=frame[y:y+h, x:x+w]
        eyes=Eye_Cascade.detectMultiScale(sub_grey, 1.1, 22)
        for (ex, ey, ew, eh) in eyes:
            cv2.rectangle(sub_frame, (ex, ey), (ex+ew, ey+eh), (255, 0, 0), 2)
        smiles=Smile_Cascade.detectMultiScale(sub_grey, 1.7, 22)
        for (sx, sy, sw, sh) in smiles:
            cv2.rectangle(sub_frame, (sx, sy), (sx+sw, sy+sh), (0, 255, 0), 2)
    return frame
```

## 4. Capturing Video

```python
Video_Capture=cv2.VideoCapture(0)
while True:
    _, frame=Video_Capture.read()
    grey=cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    canvas=Smile_detect(frame=frame, gray=grey)
    cv2.imshow('Video_Display', canvas)
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break
Video_Capture.release()
cv2.destroyAllWindows()
```
