```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_dat=pd.read_csv(r'D:\ML_Projects\Human_Activity_Recognition\train.csv')
train_dat.head()
```

```python
test_dat=pd.read_csv(r'D:\ML_Projects\Human_Activity_Recognition\test.csv')
test_dat.head()
```

```python
X_train=train_dat.iloc[:, :-2].values
Y_train=train_dat.iloc[:, -1].values
X_test=test_dat.iloc[:, :-2].values
Y_test=test_dat.iloc[:, -1].values
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
Y_train=le.fit_transform(Y_train)
Y_test=le.transform(Y_test)
```

```python
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
pca=PCA(n_components=310)
X_train=pca.fit_transform(X_train)
X_test=pca.transform(X_test)
```

```python
classifier=SVC(kernel='rbf', random_state=0)
classifier.fit(X_train, Y_train)
Y_pred=classifier.predict(X_test)
ac=accuracy_score(Y_test, Y_pred)
print(ac*100)
```
